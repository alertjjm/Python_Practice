```python
import numpy as np
import pandas
```

```python
_1m=[56, 54,58,52,54,54,56,54]
_2m=[64,62,60,60]
_3m=[64, 62, 62, 62, 62, 62, 64, 62, 66, 62, 64, 64,66,66,64,64]
_4m=[68,66,64,64,66,68]
```

```python
_1msigma=np.std(_1m)
_2msigma=np.std(_2m)
_3msigma=np.std(_3m)
_4msigma=np.std(_4m)
```

```python
import math
```

```python
from scipy.ndimage import gaussian_filter1d
import matplotlib.pyplot as plt

def gausian(x, sig, title):
    y3 = gaussian_filter1d(x, 3)
    ym = gaussian_filter1d(x,sig)
    plt.title(title)
    plt.plot(x, 'k', label='original data')
    plt.plot(y3, '--', label='filtered, sigma=3')
    plt.plot(ym, ':', label='filtered, sigma=calculated')
    plt.legend()
    plt.grid()
    plt.show()
    return ym
```

```python
_1mlist=gausian(_1m, _1msigma,'1m')
print(-np.mean(_1mlist))
_2mlist=gausian(_2m,_2msigma,'2m')
print(-np.mean(_2mlist))
_3mlist=gausian(_3m,_3msigma,'3m')
print(-np.mean(_3mlist))
_4mlist=gausian(_4m,_4msigma,'4m')
print(-np.mean(_4mlist))
```

```python
def filegaussian(filename):
    sleeparr=[]
    f = open(filename, 'r')
    while True:
        line = f.readline()
        if not line:
            break
        sleeparr.append(-int(line))
    f.close()
    resultlist=gausian(sleeparr, np.std(sleeparr), filename)
    print(np.mean(resultlist))
```

```python
filegaussian('nexus0.txt')
filegaussian('nexus1.txt')
filegaussian('nexus2.txt')
```
