```python
from scipy.ndimage import gaussian_filter1d
import matplotlib.pyplot as plt

def gausian(x, sig, title):
    y3 = gaussian_filter1d(x, 3)
    ym = gaussian_filter1d(x,sig)
    plt.title(title)
    plt.plot(x, 'k', label='original data')
    plt.plot(y3, '--', label='filtered, sigma=3')
    plt.plot(ym, ':', label='filtered, sigma=calculated')
    plt.legend()
    plt.grid()
    plt.show()
    return ym
```

```python
def filegaussian(filename):
    sleeparr=[]
    f = open(filename, 'r')
    while True:
        line = f.readline()
        if not line:
            break
        sleeparr.append(-int(line))
    f.close()
    resultlist=gausian(sleeparr, np.std(sleeparr), filename)
    print(np.mean(resultlist))
```

```python
filegaussian('1m.txt')
filegaussian('5m.txt')
filegaussian('10m.txt')
filegaussian('15m.txt')
```
